```python
import numpy as np
import os
import glob
import pandas as pd
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import plotly.graph_objects as go
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib as mpl
mpl.rcParams['axes.linewidth'] = 1.4 #set the value globally
```

```python
def mean_values(N,dim,alpha_a,alpha_g):
    #path_d = f"../../data/N_{N}/dim_{dim}/alpha_a_{alpha_a}_alpha_g_{alpha_g}"
    path_d = f"../../data_teste/N_{N}/dim_{dim}/alpha_a_{alpha_a}_alpha_g_{alpha_g}"
    all_files = glob.glob(os.path.join(path_d,"*.csv"))

    shortest_path_mean = 0
    diamater_mean = 0
    assortativity_mean = 0

    shortest_path_mean_list = []
    diamater_mean_list = []
    assortativity_mean_list = []
    
    
    num_samples = len(all_files)

    for file in all_files:
        train = pd.read_csv(file)
        shortest_path_mean_list.append(train["#mean shortest path"].values[0])
        diamater_mean_list.append(train["# diamater"].values[0])
        assortativity_mean_list.append(train["#assortativity coefficient"].values[0])
        # shortest_path_mean += train["#mean shortest path"].values[0]
        # diamater_mean += train["# diamater"].values[0]
        # assortativity_mean += train["#assortativity coefficient"].values[0]
    
    shortest_path_mean = sum(shortest_path_mean_list)/num_samples
    diamater_mean = sum(diamater_mean_list)/num_samples
    assortativity_mean = sum(assortativity_mean_list)/num_samples

    error_shortest_path_mean = np.std(shortest_path_mean_list)/np.sqrt(num_samples)
    error_diamater_mean = np.std(diamater_mean_list)/np.sqrt(num_samples)
    error_assortativity_mean = np.std(assortativity_mean_list)/np.sqrt(num_samples)


    return shortest_path_mean, diamater_mean, assortativity_mean, error_shortest_path_mean, error_diamater_mean, error_assortativity_mean
```

```python
def all_data_mean(N, dim, alpha_a, alpha_g, Alpha_A, N_aug):
    if(Alpha_A==False and N_aug==False):
        short = [[] for _ in range(len(alpha_g))]
        dia = [[] for _ in range(len(alpha_g))]
        coefAss = [[] for _ in range(len(alpha_g))]
        err_short = [[] for _ in range(len(alpha_g))]
        err_dia = [[] for _ in range(len(alpha_g))]
        err_coefAss = [[] for _ in range(len(alpha_g))]

        
        for i in range(len(alpha_g)):
            short[i].append(mean_values(N,dim,alpha_a,alpha_g[i])[0])
            dia[i].append(mean_values(N,dim,alpha_a,alpha_g[i])[1])
            coefAss[i].append(mean_values(N,dim,alpha_a,alpha_g[i])[2])
            err_short[i].append(mean_values(N,dim,alpha_a,alpha_g[i])[3])
            err_dia[i].append(mean_values(N,dim,alpha_a,alpha_g[i])[4])
            err_coefAss[i].append(mean_values(N,dim,alpha_a,alpha_g[i])[5])
        
        return short, dia, coefAss, err_short, err_dia, err_coefAss
    elif(Alpha_A==True and N_aug==False):
        shortAa = [[] for _ in range(len(alpha_a))]
        diaAa = [[] for _ in range(len(alpha_a))]
        coefAssAa = [[] for _ in range(len(alpha_a))]
        err_shortAa = [[] for _ in range(len(alpha_a))]
        err_diaAa = [[] for _ in range(len(alpha_a))]
        err_coefAssAa = [[] for _ in range(len(alpha_a))]
        
        for i in range(len(alpha_a)):
            shortAa[i].append(mean_values(N, dim, alpha_a[i], alpha_g)[0])
            diaAa[i].append(mean_values(N, dim, alpha_a[i], alpha_g)[1])
            coefAssAa[i].append(mean_values(N, dim,alpha_a[i], alpha_g)[2])
            err_shortAa[i].append(mean_values(N,dim,alpha_a[i],alpha_g)[3])
            err_diaAa[i].append(mean_values(N,dim,alpha_a[i],alpha_g)[4])
            err_coefAssAa[i].append(mean_values(N,dim,alpha_a[i],alpha_g)[5])
        return shortAa,diaAa,coefAssAa, err_shortAa ,err_diaAa, err_coefAssAa
    
    elif(Alpha_A==True and N_aug==True):
        shortN = [[] for _ in range(len(alpha_a))]
        diaN = [[] for _ in range(len(alpha_a))]
        coefAssN = [[] for _ in range(len(alpha_a))]
        err_shortN = [[] for _ in range(len(alpha_a))]
        err_diaN = [[] for _ in range(len(alpha_a))]
        err_coefAssN = [[] for _ in range(len(alpha_a))]
        for i in range(len(alpha_a)):
            for j in N:
                shortN[i].append(mean_values(j, dim, alpha_a[i], alpha_g)[0])
                diaN[i].append(mean_values(j, dim, alpha_a[i], alpha_g)[1])
                coefAssN[i].append(mean_values(j, dim,alpha_a[i], alpha_g)[2])
                err_shortN[i].append(mean_values(j, dim,alpha_a[i], alpha_g)[3])
                err_diaN[i].append(mean_values(j, dim,alpha_a[i], alpha_g)[4])
                err_coefAssN[i].append(mean_values(j, dim,alpha_a[i], alpha_g)[5])
        return shortN, diaN, coefAssN, err_shortN, err_diaN, err_coefAssN
```
