```python
# Return: dataframe with all samples

def all_properties_dataframe(N,dim,alpha_a,alpha_g):
    # Directory with all samples
    path_d = f"../../data/N_{N}/dim_{dim}/alpha_a_{alpha_a}_alpha_g_{alpha_g}"
    # dataframe with all samples
    new_file = "/properties_set.txt"
    
    # Check if directory exist
    conditional_ = os.path.exists(path_d)
    if(conditional_ == True):
        pass
    else:
        return "data doesn't exist, run code in c++ to gen data"
    
    # Check if file exist
    check_file = os.path.isfile(path_d+new_file)
    
    # Open all files path in directory .csv
    all_files = glob.glob(os.path.join(path_d,"*.csv"))
    # If file exist, open
    if check_file==True:
        df = pd.read_csv(path_d+new_file,sep=" ")
        data = df.iloc[:,1:]
        #filter_list to check if files are in dataframe
        filter_list = str(data["#cod_file"].values) 
        num_samples = len(data)
        for file in all_files:
            # Check if file are in dataframe
            conditional = os.path.basename(file)[5:-4] in filter_list
            # Make nothing if True conditional
            if(conditional==True):
                pass
            # Add new elements in dataframe
            else:
                new_data = pd.read_csv(file)
                data.loc[num_samples,"#short_path"] = new_data["#mean shortest path"].values[0]
                data.loc[num_samples,"#diamater"] = new_data["# diamater"].values[0]
                data.loc[num_samples,"#ass_coeff"] = new_data["#assortativity coefficient"].values[0]
                data.loc[num_samples,"#cod_file"] = os.path.basename(file)[5:-4]
                num_samples += 1
        # Save new dataframe update
        data.to_csv(path_d+new_file,sep=' ')

    # Else, create it
    else:
        df = pd.DataFrame(columns=["#short_path", "#diamater", "#ass_coeff","#cod_file"])
        i = 0
        # Open all files path in directory .csv
        for file in all_files:
            train = pd.read_csv(file)
            df.loc[i,"#short_path"] = train["#mean shortest path"].values[0]
            df.loc[i,"#diamater"] = train["# diamater"].values[0]
            df.loc[i,"#ass_coeff"] = train["#assortativity coefficient"].values[0]
            df.loc[i,"#cod_file"] = os.path.basename(file)[5:-4]
            i += 1
        df.to_csv(path_d+new_file,sep=' ')
```

```python
# Return: mean of all properties with err

def mean_properties(N,dim,alpha_a,alpha_g):
    # Directory with all samples
    path_d = f"../../data/N_{N}/dim_{dim}/alpha_a_{alpha_a}_alpha_g_{alpha_g}" 
    # dataframe with all samples
    new_file = "/properties_set.txt"                                           
    # Check if file exist
    check_file = os.path.isfile(path_d+new_file)

    if check_file==True:
        df = pd.read_csv(path_d+new_file, sep=' ')
        data = data = df.iloc[:,1:]
        mean_short = data["#short_path"].mean()
        mean_dia = data["#diamater"].mean()
        mean_ass = data["#ass_coeff"].mean()

        err_short = data["#short_path"].std(ddof=0)
        err_dia = data["#diamater"].std(ddof=0)
        err_ass = data["#ass_coeff"].std(ddof=0)
        return [mean_short, mean_dia, mean_ass, err_short, err_dia, err_ass]
    else:
        return [f"dataframe doesn't exist, run all_properties_dataframe(N,dim,alpha_a,alpha_g) with parameters"]
```

```python
def linear_regression(k, pk, num_imp):
    a_list = []
    b_list = []
    x = k.reshape((-1,1))
    y = pk
    model = LinearRegression()
    for i in range(num_imp):
        X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.50)
        model.fit(X_train, y_train)
        a_list.append(model.coef_)
        b_list.append(model.intercept_)
    a = np.mean(a_list)
    b = np.mean(b_list)
    err_a = np.std(a_list)
    err_b = np.std(b_list)
    return a, b, err_a, err_b
```

```python
def linear_regression_alldata(x, y, alpha_g, alpha_a, N, num_imp, Alpha_A, N_aug):
    if(Alpha_A==False and N_aug==False):
        A = [[] for _ in range(len(alpha_g))]
        B = [[] for _ in range(len(alpha_g))]
        err_A = [[] for _ in range(len(alpha_g))]
        err_B = [[] for _ in range(len(alpha_g))]

        for i in alpha_g:
            A[i].append(linear_regression(x,y,num_imp)[0])
            B[i].append(linear_regression(x,y,num_imp)[1])
            err_A[i].append(linear_regression(x,y,num_imp)[2])
            err_B[i].append(linear_regression(x,y,num_imp)[3])
        
        return A,B, err_A, err_B
    
    elif(Alpha_A==True and N_aug==False):
        A = [[] for _ in range(len(alpha_a))]
        B = [[] for _ in range(len(alpha_a))]
        err_A = [[] for _ in range(len(alpha_a))]
        err_B = [[] for _ in range(len(alpha_a))]

        for i in len(alpha_a):
            A[i].append(linear_regression(x,y,num_imp)[0])
            B[i].append(linear_regression(x,y,num_imp)[1])
            err_A[i].append(linear_regression(x,y,num_imp)[2])
            err_B[i].append(linear_regression(x,y,num_imp)[3])

        return A, B, err_A,err_B
    
    elif(Alpha_A==True and N_aug==True):
        A = [[] for _ in range(len(N))]
        B = [[] for _ in range(len(N))]
        err_A = [[] for _ in range(len(N))]
        err_B = [[] for _ in range(len(N))]
        for i in range(len(alpha_a)):
            for j in N:
                A[i].append(linear_regression(x,y,num_imp)[0])
                B[i].append(linear_regression(x,y,num_imp)[1])
                err_A[i].append(linear_regression(x,y,num_imp)[2])
                err_B[i].append(linear_regression(x,y,num_imp)[3])
        return A, B, err_A,err_B
```

```python
def linear_plot(a,b,x):
    return a*x+b
```

```python
def q(alpha_a,d):
    ration = alpha_a/d
    if(0 <= ration <= 1):
        return 4/3
    else:
        return (1/3)*np.exp(1-ration)+1
```

```python
def Lambda(alpha_a,d):
    ration = alpha_a/d
    if(0 <= ration <= 1):
        return 1/0.3
    else:
        return 1/(-1.15*np.exp(1-ration)+1.45)
```

```python
def kappa(alpha_a,d):
    ration = alpha_a/d
    if(0 <= ration <= 1):
        return 0.3
    else:
        return -1.15*np.exp(1-ration)+1.45
```
