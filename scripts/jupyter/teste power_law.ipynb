```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def pareto(alpha,xm,x):
    return alpha*(xm**alpha/x**(alpha+1))
```

```python
def power_random(rmax,alpha,num_samples):
    a = np.random.uniform(0,1,num_samples)
    if(alpha==0):
        return rmax**a
    else:
        return rmax/((a*(1-rmax**(alpha)) + rmax**alpha)**(1/alpha))
```

```python
def inverse(x,alpha):
    return x**(-alpha)
```

```python
x_l = np.linspace(1,10000,10000)
data_norm = []
alpha_a = [1.0,1.1,1.2,1.3,1.4]
labels = [f"alpha_a = {i}" for i in alpha_a]
for i in alpha_a:
    ter = inverse(x_l,i)/sum(inverse(x_l,i))
    data_norm.append(ter)    
plt.figure(figsize=(16,8))
for i in range(len(alpha_a)):
    plt.plot(x_l,data_norm[i],label=labels[i])

plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.show()
```

```python
from decimal import Decimal
diff = []
for i in range(len(alpha_a)):
    ter = max(data_norm[i])/min(data_norm[i])
    diff.append(ter)
diff = ["{:.2E}".format(Decimal(i)) for i in diff]
print(diff)
```

```python
def sa_rand(rmax,alpha,dim,num_samples):
    a = np.random.uniform(0,1,num_samples)
    alpha_dim = alpha+dim
    return (1-a*(1-rmax**(-alpha_dim+1)))**(1/(1-alpha_dim))
```

```python
import numpy as np

def generate_random_numbers_with_normalized_distribution(num_samples, r_min, r_max):
    random_uniform = np.random.uniform(0, 1, num_samples)
    random_numbers = np.exp(np.log(r_min) + random_uniform * (np.log(r_max) - np.log(r_min)))
    return random_numbers

# Set the range and number of samples
r_min = 1  # Minimum value of r
r_max = 100000  # Maximum value of r
num_samples = 500  # Number of random numbers to generate

# Generate random numbers with the desired normalized distribution
random_numbers = generate_random_numbers_with_normalized_distribution(num_samples, r_min, r_max)
```
